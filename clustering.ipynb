# Transform Data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Bigdata/data/Chemical Composion of Ceramic.csv")
```

```python
df.head()
```

```python
data = np.array(df.values[0:100])
print(data)
```

```python
X = data[:,-1]
Y = data[:,0] 
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
print(Y)
```

```python
X1 = np.array([X])
Y1 = np.array([Y])
```

```python
data = np.concatenate((X1.T,Y1.T), axis=1)
print(data)
```

# Model

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
centroid = kmeans.cluster_centers_
labels = kmeans.labels_

print (centroid)
print(labels)
```

```python
colors = ["g.","r.","c."]

for i in range(len(data)):
   print ("coordinate:" , data[i], "label:", labels[i])
   plt.plot(data[i][0],data[i][1],colors[labels[i]],markersize=10)

plt.scatter(centroid[:,0],centroid[:,1], marker = "x", s=150, linewidths = 5, zorder =10)

plt.show()
```
